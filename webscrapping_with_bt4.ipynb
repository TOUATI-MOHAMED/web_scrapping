```
import bs4 as bs 
import urllib.request

url = urllib.request.urlopen("https://www.utm.my/").read()
```

```
soup = bs.BeautifulSoup(url,'lxml')
```

```
#find all paragraphs 
#we can use this method soup.findAll('p')
soup.find_all('p')
```

```
soup.findAll('ul')
```

```
#list of paragraphs tags 
parag=soup.findAll('p')
for p in parag:
    print(p.text)
```

```
 #find all anchors  tags ,   not only tag we get the location of the links ;) 
for url in soup.findAll('a'):
    print(url.get('href'))
```

```
#how to get all links from url tags 
nav=soup.ul
for url in nav.find_all('a'):
    print(url.get('href'))
```

```
body= soup.body
for p in body.find_all('p'):
    print(p.string)
```

```
body= soup.body
for p in body.find_all('p'):
    print(p.text)
```

```
body.find_all('div',class_="container clearfix")  #we must replace the class to class_  beacuse class is a predefined name in python                                                                                                                         
```

```
for div in body.find_all('div',class_="container clearfix"):
    print(div.text)
```

```
div= body.find_all('div',class_="container clearfix") #return list 
print(div)
```

```
div= body.find('div',class_="container clearfix") #return tags into div 
print(div)
```
